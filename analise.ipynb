Importação das bibliotecas

```
import pandas as pd
import matplotlib.pyplot as plt
```

Leitura dos dados de notificações de Síndrome Gripal

```
dataframe = pd.read_csv(root_path+'/dataset/dados-pa.csv', sep=";", encoding="latin1")
```

**Dados**

```
dataframe
```

Extraindo informações de crianças e adolescentes que obtiveram resultado positivo em testes de COVID-19

```
menorIdade=0
crianca=0
list_crianca=[]
for linha in dataframe.itertuples():
  if(linha.idade<18 and linha.resultadoTeste=="Positivo"):
    menorIdade+=1
  if (linha.idade<=14 and linha.resultadoTeste=="Positivo"):
    crianca+=1
    list_crianca.append(linha.idade)
```

```
print("Quantidade de pessoas com resultado positivo menores de idade: {}".format(menorIdade))
print("Quantidade de crianças com resultado positivo: {}".format(crianca))
```

```
idade=dataframe['idade']

plt.figure(figsize=(15, 6))
plt.hist(idade, bins=range(10, 110,10))
plt.title('Distribuição de todas as idades')
plt.xlabel('Idade')
plt.ylabel('Pacientes')
```

```
plt.figure(figsize=(15, 6))
plt.hist(list_crianca)
plt.title('Distribuição de pacientes com idade igual ou menor do que 14 anos')
plt.xlabel('Idade')
plt.ylabel('Pacientes')
```
